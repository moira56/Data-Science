```python
# Phase 2: Data Analysis & Cleaning – Conclusion
```

```python
# Connecting to the Database
# Importing necessary libraries for working with the database and data
import sqlite3
import pandas as pd

# Connecting to the SQLite database
conn = sqlite3.connect("data/2019/2019-01.sqlite")

# Loading the trip data from the database, limiting to the first 10,000 rows
query = "SELECT * FROM tripdata LIMIT 10000;"
df_trips = pd.read_sql(query, conn)
```

```python
# Check for missing values
df_trips.isnull().sum()
```

```python
# Summary statistics
df_trips.describe()
```

```python
# Convert datetime columns
df_trips["tpep_pickup_datetime"] = pd.to_datetime(df_trips["tpep_pickup_datetime"])
df_trips["tpep_dropoff_datetime"] = pd.to_datetime(df_trips["tpep_dropoff_datetime"])
```

```python
# Drop congestion_surcharge column (it contains only NaN values)
df_trips = df_trips.drop(columns=["congestion_surcharge"])
```

```python
# Compute trip duration in minutes
df_trips["trip_duration"] = (df_trips["tpep_dropoff_datetime"] - df_trips["tpep_pickup_datetime"]).dt.total_seconds() / 60
```

```python
# Extract hour, day, and month from pickup time
df_trips["pickup_hour"] = df_trips["tpep_pickup_datetime"].dt.hour
df_trips["pickup_day"] = df_trips["tpep_pickup_datetime"].dt.day
df_trips["pickup_month"] = df_trips["tpep_pickup_datetime"].dt.month
```

```python
# Compute fare per mile
df_trips["fare_per_mile"] = df_trips["fare_amount"] / df_trips["trip_distance"]

# Replace infinite values with 0
df_trips.loc[df_trips["fare_per_mile"] == float("inf"), "fare_per_mile"] = 0
df_trips.loc[df_trips["fare_per_mile"] == -float("inf"), "fare_per_mile"] = 0

```

```python
# Identify if the trip was on a weekend
df_trips["is_weekend"] = df_trips["tpep_pickup_datetime"].dt.weekday >= 5
```

```python
# Categorizing fare amounts into groups
def categorize_fare(fare):
    if fare < 10:
        return "Short trip"
    elif 10 <= fare < 30:
        return "Medium trip"
    elif 30 <= fare < 60:
        return "Long trip"
    else:
        return "Premium trip"

df_trips["fare_category"] = df_trips["fare_amount"].apply(categorize_fare)
```

```python
# Display first rows with new features
df_trips.head()
```

```python
import matplotlib.pyplot as plt

# Histogram of trip duration (filtering out extreme values)
plt.figure(figsize=(8,5))
df_trips[df_trips["trip_duration"] < 60]["trip_duration"].hist(bins=50)
plt.xlabel("Trip Duration (minutes)")
plt.ylabel("Number of Trips")
plt.title("Distribution of Trip Duration")
plt.show()
```

```python
# Display the shortest and longest trips
print("Shortest trips:")
print(df_trips[["tpep_pickup_datetime", "tpep_dropoff_datetime", "trip_duration"]].nsmallest(5, "trip_duration"))

print("\nLongest trips:")
print(df_trips[["tpep_pickup_datetime", "tpep_dropoff_datetime", "trip_duration"]].nlargest(5, "trip_duration"))
```

```python
Conclusion
No missing values except congestion_surcharge, which was removed.
Feature engineering completed (trip duration, fare per mile, time-based features, fare categories).
Most trips last under 30 minutes, but extreme outliers exist.
Unrealistic trips (0 min or over 24 hours) detected and need to be removed.
```
